```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the file path to the historical dataset
file_path = "data/seoul_air_1988_2021.csv"

# Load the CSV file into a pandas DataFrame
print("Loading dataset...")
df = pd.read_csv(file_path)

# Display the first few rows to verify successful ingestion
df.head()
```

```python
# Extract the subset for the target year (e.g., 2021)
target_year = '2021'

# Safely convert to string and check if it STARTS WITH the target year
print(f"Filtering data strictly for the year {target_year}...")
df_year = df[df['dt'].astype(str).str.startswith(target_year)]

# Verify the size reduction
print(f"Original dataset rows: {len(df)}")
print(f"Filtered dataset rows: {len(df_year)}")

# Save the subset to a new CSV file
output_path = f"data/seoul_{target_year}_sample.csv"
df_year.to_csv(output_path, index=False)

#Dsiplay the first few rows of the filtered dataset
df_year.head()
```

```python
# Inspect the newly created 1-year DataFrame
print("DataFrame Shape (Rows, Columns):", df_year.shape)

# Check data types and detect missing values (Nulls)
print("\nData Types and Missing Values Info")
df_year.info()

# Display the first rows to visually confirm the subset
df_year.head()
```

```python
# Remove rows with missing data
df_clean = df_year.dropna()

y = df_clean['pm2.5']
X = df_clean.drop(columns=['pm2.5'])
```

```python
df_year = df_year.copy()

# Convert the 'dt' column to a proper datetime format
df_year['dt'] = pd.to_datetime(df_year['dt'], format='%Y%m%d%H')

print(df_year['dt'].head())
print(df_year['dt'].tail())
```

```python
numeric_df = df_year.select_dtypes(include=['float64', 'int64'])

top_correlations = numeric_df.corr()['pm2.5'].sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))

sns.barplot(x=top_correlations.values, y=top_correlations.index, palette='viridis')

plt.title("Top 10 Features Most Correlated with PM2.5", fontsize=14)
plt.xlabel("Correlation Coefficient (1.0 is a perfect match)", fontsize=12)
plt.ylabel("Features", fontsize=12)

plt.show()
```

```python
# Select the most important pollutants 
selected_features = ['so2', 'no2', 'co', 'o3', 'pm10']
X_selected = X[selected_features]
```

```python
# Scale the numbers to be on the same range
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Training features (X_train): {X_train.shape}")
print(f"Testing features (X_test): {X_test.shape}")

# Save TRAIN and TEST sets seperately
# Convert the scaled arrays back into pandas DataFrames with column names
df_X_train = pd.DataFrame(X_train, columns=selected_features)
df_X_test = pd.DataFrame(X_test, columns=selected_features)

# Add the target variable (y) back to each set
df_train = df_X_train.copy()
df_train['pm2.5'] = y_train.values

df_test = df_X_test.copy()
df_test['pm2.5'] = y_test.values

# Save the training and testing datasets as separate CSV files
train_file = "data/seoul_2021_train.csv"
test_file = "data/seoul_2021_test.csv"

df_train.to_csv(train_file, index=False)
df_test.to_csv(test_file, index=False)

print(f"\nSuccessfully saved perfectly split data:")
print(f"- {train_file}")
print(f"- {test_file}")
```
